```python
import pandas as pd
```

```python
data = pd.read_csv("all_searched_entities_26_08_24.tsv", sep='\t')
```

```python
data
```

```python
relationships = []
for a, b, c in zip(data['ENTITYA'], data['EFFECT'], data['ENTITYB']):
    relationships.append((a, b, c))
```

```python
relationships
```

```python
import networkx as nx
graph = nx.DiGraph();
```

```python
for relationship in relationships:
    graph.add_edge(relationship[0], relationship[2], label=relationship[1])
```

```python
import matplotlib.pyplot as plt
from pyvis.network import Network

G_viz = Network(height='1000px', width='100%', bgcolor='white', notebook=True, 
font_color ='black')
G_viz.from_nx(graph)
G_viz.toggle_physics(True);
graph_name = 'base_graph'
G_viz.show(graph_name + '.html', notebook=False)

options = {
        "interaction":{
            "hideEdgesOnDrag": True,
            "hideEdgesOnZoom": True,

        },
          "nodes":{
              "font":{
                  "size": 50,
                  "bold":True
              }
          },
          "edges":{
              "color":'red',
              "smooth":False,
          },
          "physics":{
              "barnesHut":{
                  "gravitationalConstant":-500000,
                  "centralGravity":12,
                  "springLength": 50,
                  "springConstant": 0.7,
                  "damping": 3,
                  "avoidOverlap": 10
              }
          },
}

G_viz.options=options

# saving the graph as a gml file
graph_file = graph_name + '.gml'
nx.write_gml(graph, graph_file)
```
